```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
# this dataset has the global confirmed cases of COVID-19 
# from january 22,2020 to July 15,2020

corona_dataset_confirmed = pd.read_csv("time_series_covid19_confirmed_global.csv")

corona_dataset_confirmed.head(10)

```

```python
# deleteing lat and long. no need 


corona_dataset_confirmed.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
corona_dataset_confirmed.describe(include="all")
```

```python
# need to aggregate the data
# aggregating the data by row and country
# from january 22,2020 to July 15,2020

corona_dataset_confirmed_aggregated = corona_dataset_confirmed.groupby("Country/Region").sum()
corona_dataset_confirmed_aggregated.head(10)
```

```python
# calculating the max infection rate by each county

countries=list(corona_dataset_confirmed_aggregated.index)
max_infection_rate=[]
for c in countries:
    max_infection_rate.append(corona_dataset_confirmed_aggregated.loc[c].diff().max())
    
max_infection_rate

corona_dataset_confirmed_aggregated["max_infection_rate"] = max_infection_rate
corona_dataset_confirmed_aggregated.head(14)

#corona_dataset_confirmed_aggregated.loc["Bangladesh"]["max_infection_rate"]
```

```python
#load dataset global death and aggragate

corona_dataset_death = pd.read_csv("time_series_covid19_deaths_global.csv")

corona_dataset_death.drop(["Lat","Long"],axis=1,inplace=True)
corona_dataset_death_aggregated = corona_dataset_death.groupby("Country/Region").sum()

corona_dataset_death_aggregated.head(10)
```

```python
countries_death=list(corona_dataset_death_aggregated.index)
max_death_rate=[]
for c in countries_death:
    max_death_rate.append(corona_dataset_death_aggregated.loc[c].diff().max())
    


corona_dataset_death_aggregated["max_death_rate"] = max_death_rate
corona_dataset_death_aggregated.head(14)
```

```python
#load dataset global recovered and aggragate

corona_dataset_recovered = pd.read_csv("time_series_covid19_recovered_global.csv")

corona_dataset_recovered.drop(["Lat","Long"],axis=1,inplace=True)
corona_dataset_recovered_aggregated = corona_dataset_recovered.groupby("Country/Region").sum()

corona_dataset_recovered_aggregated.head(10)

countries_recovered=list(corona_dataset_recovered_aggregated.index)
max_recovered_rate=[]
for c in countries_recovered:
    max_recovered_rate.append(corona_dataset_recovered_aggregated.loc[c].diff().max())
    


corona_dataset_recovered_aggregated["max_recovered_rate"] = max_recovered_rate
corona_dataset_recovered_aggregated.head(14)
```

```python
# Visulaize all correlations max_infection_rate

col_dtypes = data.dtypes.tolist()
numeric_cols = [data.columns[idx] for idx, col_type in enumerate(col_dtypes) if col_type=='int64' or col_type=='float64']
df_numerics = data[numeric_cols]
f, axes = plt.subplots(3,3)
f.set_figheight(15)
f.set_figwidth(25)
col_idx = 0
for i in range(3):
    for j in range(3):
        x = numeric_cols[col_idx]
        sns.regplot(x, y="max_infection_rate", data=df_numerics, ax=axes[i][j])
        col_idx += 1
```
